# Copies and Views

```python
a = np.array(range(10)) #bad way :)
```

```python
a
```

```python
b = a[::2]
```

```python
b
```

```python
np.shares_memory(a,b)
```

```python
b
```

```python
b[::] = 100
```

```python
b
```

```python
a
```

```python
a = np.arange(10) # Good way
```

```python
a 
```

```python
c = a[::2].copy()
```

```python
a
```

```python
c
```

```python
np.shares_memory(a,c)
```

```python
c[::] = 200
```

```python
c
```

```python
a
```

# Fancy Indexing

```python
a = np.random.randint(1, 5, 10)
```

```python
a
```

```python
a % 2 == 0
```

```python
mask = (a % 2 == 0)
```

```python
mask
```

```python
a[mask]
```

```python
a[mask] = -10
```

```python
a
```
